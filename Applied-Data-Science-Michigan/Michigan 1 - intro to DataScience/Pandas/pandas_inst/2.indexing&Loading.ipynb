## Boolean Masking



```python
df = pd.read_csv(
    'NoteBooks/resources\week-2\datasets\Admission_Predict.csv',index_col=0)
df.columns = (x.lower().strip() for x in df.columns)
df.head()
```

```python
admit_mask = df['chance of admit'] > 0.7
admit_mask
```

```python
df.where(admit_mask).head()
df.where(admit_mask, np.nan).head()
df.where(admit_mask, np.nan).dropna().head()
```

```python
df[df['chance of admit'] > 0.7].head()
```

```python
# * we can't use opreators like and / or while using mask
# * so we use & / | to get the mask

(df['chance of admit']>0.7) | (df['gre score'] > 320)
(df['chance of admit']>0.7) & (df['gre score'] > 320)

#* make sure to put each comparison in parentheses
```

```python
df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)

df[df['chance of admit'].gt(0.7) & df['chance of admit'].lt(0.9)]
#* gt : greater than
#* lt : less than
#* no parentheses needed
df[(df['toefl score'].isin(range(99,120)))]
#* or

#df[df['chance of admit'].gt(0.7).lt(0.9)]
```
