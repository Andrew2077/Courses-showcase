```python
df2 = pd.read_csv('NoteBooks/resources\week-3\datasets\listings.csv')
df2.head(1)
print(df2['cancellation_policy'].unique())
print(df2['review_scores_value'].unique())
```

```python
df2 = df2.set_index(['cancellation_policy', 'review_scores_value'])
```

```python
for group, frame in df2.groupby(level=(0, 1)):
    print(group)
```

```python
def grouping_tens(item):
    if item[1] == 10.0:
        return (item[0], "10.0")
    else:
        return (item[0], "Not 10.0")
    
for group, frame in df2.groupby(by= grouping_tens):
    print(group)
```

```python
def felx_only(item):
    if item[0] == 'flexible':
        return (item[1], "flexible")
    
for group, frame in df2.groupby(by= felx_only):
    print(group)
```

## Applying

```python
df = pd.read_csv('NoteBooks/resources\week-3\datasets\listings.csv')
df = df[['cancellation_policy', 'review_scores_value']]
df.head()
```

```python
def normalize(group):
    avg = np.nanmean(group['review_scores_value'])
    group['normalized_review_score'] = np.absolute(avg- group['review_scores_value'])
    return group
    
    #* mean of the group - moderate , flexible etc..
```

```python
df.groupby('cancellation_policy').apply(normalize).dropna().head().drop('Normalized_scire', axis=1)
```

```python
df['Normalized_score'] = df['review_scores_value'].apply(lambda x: np.absolute(np.nanmean(df['review_scores_value']) - x))

df[['cancellation_policy', 'review_scores_value', 'Normalized_score']].dropna().head()

#* total mean of all groups
```
