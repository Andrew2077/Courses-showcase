```python
import pandas as pd
import numpy as np


```

```python
group = {'name': ['Anastasia', 'Dima', 'Katherine', 'James',
                  'Emily', ],
         'role ': ['student', 'teacher', 'manager', 'student', 'couder'],
         }

info = {'name': ['mandy', 'dema','Katherine', 'James',
                 'Emily', ],
        'salary': [120000, 80000, 100000, 90000, 80000],
        }

group = pd.DataFrame(group)
info = pd.DataFrame(info)


```

## pd.merge Merges on columns

```python
pd.merge(group, info, how= 'outer')
```

```python
pd.merge(group, info, how= 'inner').set_index('name')
```

```python
pd.merge(group, info, how= 'right')
```

```python
pd.merge(group, info, how= 'cross').head()
```

## pd.concat merges on both columns or rows

```python
pd.concat([group, info], axis=0, keys=['group', 'info'])
#* axis=0 for vertical concatination
#* axis=1 for horizontal concatination
```
