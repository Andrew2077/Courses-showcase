```python
import pandas as pd
import numpy as np

df = pd.read_csv('datasets/cwurData.csv')
```

```python
df.head()
```

```python
def classify_rank(group):
    if group < 100:
        rank = 'gold'
        return rank
    elif group< 200:
        rank = 'silver'
        return rank
    elif group< 300:
        rank = 'bronze'
        return rank
    else:
        rank = 'iron'
        return rank
    
rank = df['world_rank'].apply(lambda x : classify_rank(x))
```

```python
df.insert(0, 'tier', rank)
```

```python
df.drop('tier', axis=1, inplace=True)
```

```python
df.head()
```

## Pivot tabel

```python
df.pivot_table(values='score', index='country',
               columns='tier', aggfunc=[np.mean], margins=True).head()
```

```python
df.pivot_table(values='score', index='country', columns='tier',
               aggfunc=[np.mean, np.max], margins=True).head()
```

```python
new_df = df.pivot_table(values='score', index='country', columns='tier',
               aggfunc=[np.mean, np.max], margins=True)

print(new_df.index)
print(new_df.columns)
```
