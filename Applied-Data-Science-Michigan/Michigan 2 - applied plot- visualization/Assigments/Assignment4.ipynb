```python
salaries = pd.read_csv('ds_salaries.csv')
Data_Scientist =salaries[salaries['job_title'] == 'Data Scientist'].reset_index(drop=True)
# Data_Scientist.drop(['job_title', ], axis=1, inplace=True)
Data_Scientist.drop(['Unnamed: 0','salary','salary_currency','job_title'], axis=1, inplace=True)
Data_Scientist = Data_Scientist[Data_Scientist['employment_type']=='FT']
#Data_Scientist = Data_Scientist[Data_Scientist['company_location']=='US']
Data_Scientist
```

```python
MI_sal = Data_Scientist[Data_Scientist['experience_level'] == 'MI'].groupby(
    'work_year').mean()['salary_in_usd'].round(0)
SE_sal = Data_Scientist[Data_Scientist['experience_level'] == 'SE'].groupby(
    'work_year').mean()['salary_in_usd'].round(0)
EN_sal = Data_Scientist[Data_Scientist['experience_level'] == "EN"].groupby(
    'work_year').mean()['salary_in_usd'].round(0)
```

```python
Ds_sal = pd.DataFrame({'Junior DS': EN_sal, 'Senior DS': MI_sal, 'Expert DS': SE_sal, })
```

```python
Ds_sal
```

```python
plot = Ds_sal.plot(kind='bar', stacked=False, figsize=(10, 5),xlabel='', ylabel='Salary', table=True, title= 'Salary by Experience Level in (USD)')
plot.set_xticklabels('')
```

As you can see the Avg Salary for Juniors is decreasing and the Avg Salary for Seniors is increasing. the avg salary for expers should increase but duo having not enough data to show that. it's variant
