```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib widget
```

```python
np.random.seed(1234), plt.style.use('seaborn-darkgrid')

v1 = pd.Series(np.random.normal(0, 10, size=1000), name='v1')
v2 = pd.Series(2 * v1 + np.random.normal(60, 15, size=1000), name='v2')
```

```python
plt.figure()
plt.hist(v1, alpha = 0.7, bins = np.arange(-50,150,5), label = 'v1')
plt.hist(v2, alpha = 0.7, bins = np.arange(-50,150,5), label = 'v2')
plt.legend()
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked',density=True)
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3)
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color':'#00D0F9'}, kde_kws={'color':'#FA58AC'})
```

```python
sns.jointplot(v1, v2, alpha = 0.5)
```

```python
grid = sns.jointplot(v1, v2, alpha = 0.5)
grid.ax_joint.set_aspect('equal')
```

### hexbin plot

```python
sns.jointplot(v1, v2, kind='hex', color='#4CB391')
```

### margional distribution plot

```python
sns.jointplot(v1, v2, kind='kde', space=0,);
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

scatterplot matrix in sns

```python
sns.pairplot(iris, hue='Name', palette='Set1', diag_kind='kde', size=2);
```

### violin plot

```python
plt.figure(figsize=(12,8))

plt.subplot(141)
sns.swarmplot(x='Name', y='PetalLength', data=iris, palette='Set1')

plt.subplot(142)
sns.violinplot(x='Name', y='PetalLength', data=iris, palette='Set1')


```
